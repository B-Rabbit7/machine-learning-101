# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. prep data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. set up Plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. plot the data
ax.plot(x,y)

# 4. customize

ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. save the figure
fig.savefig("../Images/test")
```

# Two options for subplots

```python
# Subplot #1

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(but_butter_prices.keys(), but_butter_prices.values())
ax4.hist(np.random.randn(10000));
```

```python
# Subpllot option 2

fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(but_butter_prices.keys(),but_butter_prices.values())
ax[1,1].hist(np.random.randn(10000));
```

### Which one should we use (pyplt vs matplotlib OO method)

* When plotting something quick, pyplot is ok
* When plotting something advanced, use OO method

```python
over_50 = heart_diseas[heart_diseas["age"]> 50]
over_50
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",x='age',y='chol',c="target");
```

```python
# OO method
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x='age',y='chol',c="target",ax=ax);
```

```python
# Set the style
plt.style.use('seaborn')

## OO method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

# Customize

ax.set(title="Heart disease and Cholesterol level", xlabel="Age", ylabel="Cholesterol");

# Add legend

ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),linestyle="-.");
```

## All in one: Subplots wit OO methods and cmap customization

```python
# subplot using OO
figure,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter") # changing the color scheme of the dots
ax0.set(title="Heart Disease and Cholesterol levels",
        xlabel="Age",
        ylabel="Cholesterol")
## Set the x-axis limit
ax0.set_xlim([50,80])

# add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(over_50["chol"].mean(),linestyle="-.");

# add stuff to ax1

scatter2 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"], cmap="spring")
ax1.set(title="Max heart rate vs Age",xlabel="Age", ylabel="Max Heart Rate")

## Change ax1 x-axis limit
ax1.set_xlim([50,80])
ax1.legend(*scatter2.legend_elements(),title="target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="-.");

figure.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
